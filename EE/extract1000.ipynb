```python
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
level1_df = pd.read_csv('results_v2/clean_results/level1/all_events.csv', sep='\t', dtype='string')
level2_df = pd.read_csv('results_v2/clean_results/level2/all_events.csv', sep='\t', dtype='string')
level3_df = pd.read_csv('results_v2/clean_results/level3/all_events.csv', sep='\t', dtype='string')
unique_event_in_level1 = level1_df[~level1_df['ID'].isin(level2_df['ID']) ]
unique_event_in_level2 = level2_df[~level2_df['ID'].isin(level3_df['ID'])]
num_level2 = len(unique_event_in_level2)
num_level1 = len(unique_event_in_level1)
num_level3 = len(level3_df)
categories = ['num of level1', 'num of level2', 'num of level3']
values = [num_level1, num_level2, num_level3]
plt.bar(categories, values)
plt.show()
```

```python
dict_id1to2 = json.load(open('dict_hier_id.json', 'r'))
dict_id2to3 = json.load(open('dict_hier_id_level23.json', 'r'))
level1_ids = []
level2_ids = []
level3_ids = []
for key, value in dict_id1to2.items():
    level1_ids.append(key)
for id in level1_ids:
    for key, value in dict_id1to2[id].items():
        level2_ids.append(key)
for id in level2_ids:
    if id in dict_id2to3:
        for ids in dict_id2to3[id]:
            level3_ids.append(ids)
```

```python
values = []
categories = []
for id in level1_ids:
    values.append((level1_df['Relation_id'] == id).sum())
    categories.append(id)
plt.bar(categories, values)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
values = []
categories = []
for id in level1_ids[0:5]:
    values.append((level1_df['Relation_id'] == id).sum())
    categories.append(id)
    for key, value in dict_id1to2[id].items():
        values.append((level2_df['Relation_id'] == key).sum())
        categories.append(key)
plt.figure(figsize=(200, 50))
plt.bar(categories, values)
plt.xticks(rotation=45, ha='right', fontsize = 60)
plt.yticks(fontsize = 60)
plt.show()
```
